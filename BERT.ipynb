```
import numpy as np
from flair.embeddings import WordEmbeddings
from flair.embeddings import TransformerWordEmbeddings
from flair.data import Sentence
from scipy.spatial import distance
```

```
glove_embedding = WordEmbeddings('glove')
```

```
sen = Sentence('apple released iphone 12 pro max in 2020')
glove_embedding.embed(sen)
```

```

for token in sen:
  print(token)
  print(token.embedding)
  print('\n')
```

```

sen1 = Sentence('an apple a day keeps doctor away')
glove_embedding.embed(sen1)
```

```
for t in sen1:
  print(t)
  print(t.embedding)
  print('\n')
```

```
# Glove does not understand the context 
glove_dist = distance.euclidean(np.array(sen[0].embedding),np.array(sen1[1].embedding))
print('Distance between apple embeddings for Glove = {}'.format(glove_dist))
```

#Best Embeddings

```
best_embedding = TransformerWordEmbeddings('bert-base-multilingual-cased')
```

```
best_embedding = TransformerWordEmbeddings('bert-base-multilingual-cased')
best_embedding.embed(sen)
for token in sen:
  print(token)
  print(token.embedding)
```

```

best_embedding.embed(sen1)
for token in sen1:
  print(token)
  print(token.embedding)
```
