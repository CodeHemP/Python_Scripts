```python
import os
import wordcloud as wc
```

```python
f = open('E:/Python/Scripts/introinsta.txt')
contents = f.read()
f.close()
words = contents.split(' ')
print(words[:25])
```

```python
words = [w for w in words if len(w)>1]
print(words[:25])
```

```python
words = [w.lower() for w in words]
print(words[:25])
```

```python
words = [w.strip() for w in words]
print(words[:25])
```

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
words = [stemmer.stem(w) for w in words]
print(words[:25])
```

```python
from collections import Counter
ctr = Counter(words)
print(ctr['ottoman'])
print(ctr.most_common(10))
```

```python
print([p[0] for p in ctr.most_common(10)])
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud()
wordcloud.fit_words(ctr)

fig = plt.figure(figsize=(5,4))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
# 30 most common word-freq pairs

wtuples = ctr.most_common(30)
wdict = dict(wtuples)

wordcloud = WordCloud()
wordcloud.fit_words(wdict)

fig = plt.figure(figsize=(5,4))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
english = list(ENGLISH_STOP_WORDS)
print(english[:25])
```

```python
goodwords = [w for w in words if w not in ENGLISH_STOP_WORDS]
goodctr = Counter(goodwords)
print(goodctr.most_common(10))
```

```python
wtuples = goodctr.most_common(30)
wdict = dict(wtuples)

wordcloud = WordCloud()
wordcloud.fit_words(wdict)

fig = plt.figure(figsize=(5,3))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
